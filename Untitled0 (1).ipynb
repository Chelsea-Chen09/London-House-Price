Price prediction

```python
price_prediction_eng = df_England.assign(year = df_England['year'].dt.year)
price_prediction_eng = price_prediction_eng.groupby('year',as_index = False).agg({'average_price': 'mean'})
price_prediction_eng
```

```python
price_prediction_ldn = df_London.assign(year = df_London['year'].dt.year)
price_prediction_ldn = price_prediction_ldn.groupby('year',as_index = False).agg({'average_price': 'mean'})
price_prediction_ldn
```

```python
years_eng = np.array(price_prediction_eng['year'])
prices_eng = np.array(price_prediction_eng['average_price'])
A = years_eng.reshape(-1,1)
a = prices_eng.reshape(-1,1)
```

```python
years_ldn = np.array(price_prediction_ldn['year'])
prices_ldn = np.array(price_prediction_ldn['average_price'])
B = years_ldn.reshape(-1,1)
b = prices_ldn.reshape(-1,1)
```

```python
model_eng = LinearRegression()
model_eng.fit(A, a)
```

```python
model_ldn = LinearRegression()
model_ldn.fit(B, b)
```

```python
future_dates_eng = np.arange(2024, 2029).reshape(-1, 1)
print(future_dates_eng)
```

```python
future_dates_ldn = np.arange(2024, 2029).reshape(-1, 1)
print(future_dates_ldn)
```

```python
forecast_eng = model_eng.predict(future_dates_eng)
forecast_eng
```

```python
forecast_ldn = model_ldn.predict(future_dates_ldn)
forecast_ldn
```

```python
forecastdic_eng = {'year':future_dates_eng.flatten(), 'average_price':forecast_eng.flatten()}
forecastdf_eng = pd.DataFrame(forecastdic_eng)
forecastdf_eng
```

```python
forecastdic_ldn = {'year':future_dates_ldn.flatten(), 'average_price':forecast_ldn.flatten()}
forecastdf_ldn = pd.DataFrame(forecastdic_ldn)
forecastdf_ldn
```

Outlier cleaning

```python
# houses sold
df_factor['houses_sold1'] = df_factor['houses_sold'].apply(lambda x: np.log(x) if (x > 1000) else np.NaN)
```

```python
# price
df_factor['average_price1'] = df_factor['average_price'].apply(lambda x: np.nan if x > 600000 else x)
```

```python
# cirme
IQR_crime = df_factor['no_of_crimes'].quantile(0.75) - df_factor['no_of_crimes'].quantile(0.25)
upper_limit = df_factor['no_of_crimes'].quantile(0.75) + IQR_crime
df_factor['no_of_crimes1'] = df_factor['no_of_crimes'].apply(lambda x: np.nan if (x > upper_limit) or (x<1000) else x)
```

```python
# mean salary
IQR_salary = df_factor['mean_salary'].quantile(0.75) - df_factor['mean_salary'].quantile(0.25)
upper_limit = df_factor['mean_salary'].quantile(0.75) + IQR_salary*0.9
df_factor['mean_salary1'] = df_factor['mean_salary'].apply(lambda x: np.nan if x > upper_limit else x)
```

```python
# population size
df_factor['population_size1'] = df_factor['population_size'].apply(lambda x: np.nan if x < 50000 else x)
```

```python
# number of jobs
df_factor['number_of_jobs1'] = df_factor['number_of_jobs'].apply(lambda x: np.log(x) if (x > 0) & (x<350000) else np.nan)
```

```python
# life satisfication no futher action
```

```python
# number of houses in the market
df_factor['no_of_houses1'] = df_factor['no_of_houses'].apply(lambda x: np.nan if x < 20000 else x)
```
